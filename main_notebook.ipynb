#### First API

##### Hecho con dataframes

```python
import requests
import json

response = requests.get('https://api.fbi.gov/wanted/v1/list')
print("FBI Status Code:", response.status_code)
results = response.json()
```

```python
import pandas as pd
data = pd.DataFrame(results)
data.head()
```

```python
data.shape
```

```python
from pandas import json_normalize

flattened_data = json_normalize(data['items'])

flattened_data.head()
```

##### Hecho con json.loads

```python
import requests
import json

response = requests.get('https://datos.madrid.es/egob/catalogo/205026-0-cementerios.json')
print("Cementerio Status Code:", response.status_code)
```

```python
# print(response.json())
```

```python
data = json.loads(response.content)
# print(data['total'])
# print(data['items'][0]['title'])
```

```python
# from pandas.io.json import json_normalize # DEPRECATED
from pandas import json_normalize

results = response.json()
flattened_data = json_normalize(data['@graph'])

flattened_data.drop("@id",axis=1,inplace=True)
flattened_data.drop("@type",axis=1,inplace=True)
flattened_data.drop("id",axis=1,inplace=True)

flattened_data.head()
```
