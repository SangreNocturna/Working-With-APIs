#### First API

```python
import requests
import json

response = requests.get('https://api.fbi.gov/wanted/v1/list')
print("FBI Status Code:", response.status_code)
```

```python
data = json.loads(response.content)
print(data['total'])
print(data['items'][0]['title'])
```

```python
import pandas as pd
data = pd.DataFrame(data)
data.head()
```

```python
data.shape
```

```python
from pandas import json_normalize

flattened_data = json_normalize(data['items'])

flattened_data.head()
```
